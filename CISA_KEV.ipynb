```python
# @title Bootstrap CISA KEV data
# load pandas library as "pd"
!pip install pandas
import pandas as pd


# Every vuln in the KEV...
# 1. has a CVE
# 2. is being actively exploited
# 3. has clear remediation guidance

# Grab CISA KEV (JSON)
kev_json = 'https://www.cisa.gov/sites/default/files/feeds/known_exploited_vulnerabilities.json'

# convert it to a data frame
df = pd.read_json(kev_json) # title / catalogVersion / dataReleased / count / vulnerabilities
```

```python
# @title Capture Vuln Entries in a Dataframe
# Prune dataframe to just the vulnerabiliites column
vuln_series = df['vulnerabilities']
vuln_df = vuln_series.to_frame()
vuln_df
```

```python
# @title Chunking Through Each Entry in the KEV
for kev_entry in df['vulnerabilities']:
  print('\n')
  for key, value in kev_entry.items():
    print(key,":", value)
```
